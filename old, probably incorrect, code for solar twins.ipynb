def find_m_b(x,y,err):
    errsq = np.square(err)
    errC = np.diag(errsq)
    ab = ([1] * len(x))
    atemp = x
    mat = []
        
    for z, txt in enumerate(atemp):
        mat.append(atemp[z])
        mat.append(ab[z])
            
    a= np.matrix(mat).reshape((len(x), 2))
    At = np.transpose(a)
    inverrC = np.linalg.inv(errC)
    pt1 = np.dot(At, np.dot(inverrC,a))
    inv1= np.linalg.inv(pt1)
    pt2 = np.dot(At, np.dot(inverrC, y)).T
        
    cov = np.dot(inv1, pt2)
        
    m_= float(cov[0])
    b_= float(cov[1])
    return m_,b_

def chisq(param, x,y, error):
    [m,b] = param
    chi = 0 
    length = 30
    myvalu = 0
    while myvalu < length :   
        fxi = (m*(x[myvalu]) + b)
        num = (y[myvalu] - fxi)**2
        den = (error[myvalu])**2
        val = num/den
        chi = chi + val
        myvalu = myvalu + 1
    return chi
