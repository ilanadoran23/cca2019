```python
mg_con_temp = []
for ob in t['star_name']:
    abund = np.array(star_table(ob)['Abundance'])
    con_temp = np.array(star_table(ob)['Condensation Temp'])
    error = np.array(star_table(ob)['Abundance Error'])
    residu = residuals(con_temp, abund, error)
    mg_con_temp.append(residu[2])

x1 = (0.00135163, -0.00619703, -0.00397238) #one-dimensional optimization
reso = scipy.optimize.minimize(nlnL, x1, args = (t['age'], mg_con_temp, t['Mg_err']))
```

```python
plt.scatter(t['age'], mg_con_temp) 
plt.xlabel('Age (Gyr)',fontsize=25)
plt.ylabel('Mg/Fe (dex)', fontsize=25)
plt.title('Mg Residual Abundance vs. Stellar Age', fontsize=25)
plt.xticks(size = 20)
plt.yticks(size = 20)

mb= reso['x']
plot_xs = np.arange(0, 9, .1)
plt.plot(plot_xs, mb[0] * plot_xs + (mb[1]), color = 'black', linewidth=2)
plt.text(8,-.06,'standard deviation = {0}'.format(stdev(mg_con_temp)),horizontalalignment='right', fontsize=20)

#error bars 
plt.errorbar(t['age'], mg_con_temp, yerr= t['Mg_err'], xerr= t['age_err'], fmt='o', color='mediumseagreen',
                 ecolor='lightsteelblue', elinewidth=1,capsize=0)

plt.savefig('mgvsageresid.png')
```

```python
table= t.copy()
table.remove_column('O')
table.remove_column('C')

elements =[]
elnames = table.columns[3:64]
for n in elnames:
    if len(n) < 3 :
        elements.append(n)
```

```python
#plot of star age vs Mg abundance from above table - with random condensation temperature removed

plt.scatter(t['age'], t['Mg'], c='firebrick') 
plt.xlabel('Age (Gyr)',fontsize=25)
plt.ylabel('Mg/Fe (dex)', fontsize=25)
plt.title('Mg Abundance vs. Stellar Age', fontsize=25)
plt.xticks(size = 20)
plt.yticks(size = 20)
        
mbtemp= restemp['x']
plot_xs = np.arange(0, 9, .1)
plt.plot(plot_xs, mbtemp[0] * plot_xs + (mbtemp[1]), color = 'black', linewidth=2)
plt.text(8,-.06,'standard deviation = {0}'.format(stdev(t['Mg'])),horizontalalignment='right', fontsize=20)

plt.errorbar(t['age'], t['Mg'], yerr= t['Mg_err'], xerr= t['age_err'], fmt='o', color='firebrick',
                 ecolor='lightsteelblue', elinewidth=1,capsize=0)
```

```python
con_temp_shuffle = []
for n in elements:
    con_temp_shuffle.append(tc_map[n])
    np.random.shuffle(con_temp_shuffle)

mg_con_temp_shuffle = []
for ob in t['star_name']:
    abund = np.array(star_table(ob)['Abundance'])
    error = np.array(star_table(ob)['Abundance Error'])
    residu = residuals(con_temp_shuffle, abund, error)
    mg_con_temp_shuffle.append(residu[2])

x0 = (0.00135163, -0.00619703, -0.00397238) #one-dimensional optimization
resshuff = scipy.optimize.minimize(nlnL, x0, args = (t['age'], mg_con_temp_shuffle, t['Sr_err']))

plt.scatter(t['age'], mg_con_temp_shuffle) 
plt.xlabel('Age (Gyr)',fontsize=25)
plt.ylabel('Mg/Fe (dex)', fontsize=25)
plt.title('Mg Residual Abundance vs. Stellar Age', fontsize=25)
plt.xticks(size = 20)
plt.yticks(size = 20)

mb= resshuff['x']
plot_xs = np.arange(0, 9, .1)
plt.plot(plot_xs, mb[0] * plot_xs + (mb[1]), color = 'black', linewidth=2)
plt.text(8,.06,'standard deviation = {0}'.format(stdev(mg_con_temp_shuffle)),horizontalalignment='right', fontsize=20)
plt.errorbar(t['age'], mg_con_temp_shuffle, yerr= t['Mg_err'], xerr= t['age_err'], fmt='o', color='k',
                 ecolor='lightsteelblue', elinewidth=1,capsize=0)
```

```python
limit = 60
mg_standarddev = []

while limit > 0:
    mg_con_temp_shuffle = []
    con_temp_shuffle = []
    for n in elements:
        con_temp_shuffle.append(tc_map[n])
        np.random.shuffle(con_temp_shuffle)
    for ob in t['star_name']:
        abund = np.array(star_table(ob)['Abundance'])
        error = np.array(star_table(ob)['Abundance Error'])
        residu = residuals(con_temp_shuffle, abund, error)
        mg_con_temp_shuffle.append(residu[2])
        
    mg_standarddev.append(stdev(mg_con_temp_shuffle))
    limit = limit - 1
```

```python
mg_standarddev
```

```python
greater = 0
less = 0
for sd in mg_standarddev:
    if sd > stdev(t['Mg']):
        greater = greater + 1
    else: 
        less = less + 1    
```
