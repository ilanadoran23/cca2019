```python
age = np.array(MgTable['Stellar Age'])
mg = np.array(MgTable['Mg Corrected Abundance'])
err = np.array(MgTable['Mg Error'])
age_err = np.array(t['age_err'])

jackk = find_m_b(age, mg, err)
s = covmatrix(age,mg,err)
```
